### Separate agg tables then merging approach

It seems hard to create multiple nlargest aggregation columns with everything else coming before it. I'm going to create multiple groupby (grouping using the same columns) tables then merge their results later

### I. Set up the index and basic numerical stats

```python
## basic setup

nat_group = ['year', 'quarter','months','case_status']
nat_agg = {
    'index': 'count', 
    'tot_workers': np.sum, 
    'yr_sal': [np.min, np.mean, np.max],
    'pw_sal': [np.min, np.mean, np.max],
    'dec_dur': (lambda x: x.mean())
}
```

#### groupby function

```python
nat_byYear = dat.groupby(['year', 'case_status']).agg(nat_agg)

nat_byYear
```

#### clean up the tabular format

```python
## Renaming columns 
nat_byYear.columns = ["_".join(x) for x in nat_byYear.columns.ravel()]
nat_byYear.columns = ['app_count', 'worker_count', 
                      'min_yr_sal', 'avg_yr_sal', 'max_yr_sal', 
                      'min_pw', 'avg_pw', 'max_pw', 
                      'avg_decdur']
```

```python
nat_byYear['index'] = nat_byYear.index
nat_byYear['year'] = nat_byYear['index'].apply(lambda x: x[0])
nat_byYear['result'] = nat_byYear['index'].apply(lambda x: x[1])
##nat_byYear['peiord'] = nat_byYear['index'].apply(lambda x: x[2])
```

```python
nat_byYear.head()
```

```python
# float values to int 
nat_byYear['min_yr_sal'] = nat_byYear['min_yr_sal'].apply(np.floor).astype(int)
nat_byYear['avg_yr_sal'] = nat_byYear['avg_yr_sal'].apply(np.floor).astype(int)
nat_byYear['max_yr_sal'] = nat_byYear['max_yr_sal'].apply(np.floor).astype(int)
nat_byYear['min_pw'] = nat_byYear['min_pw'].apply(np.floor).astype(int)
nat_byYear['avg_pw'] = nat_byYear['avg_pw'].apply(np.floor).astype(int)
nat_byYear['max_pw'] = nat_byYear['max_pw'].apply(np.floor).astype(int)

# round to 'days'
nat_byYear['avg_decdur'] = nat_byYear['avg_decdur'].dt.days
```

```python
len(nat_byYear)
```

### II. Industry Code and Job Title Freq

#### top industry indicators: naics code, job titles

```python
dat.dtypes
```

```python
nat_naics5 = dat.groupby(['year', 'case_status'])['naics5'].value_counts().groupby(['year', 'case_status']).nlargest(5).to_frame()
#nat_naics5.groupby(['year','case_status']).agg({'naics5': lambda x: ', '.join(x)})
#nat_naics5.columns = ["_".join(x) for x in nat_naics5.columns.ravel()]

nat_naics5['index'] = nat_naics5.index
nat_naics5['year'] = nat_naics5['index'].apply(lambda x: x[0])
nat_naics5['result'] = nat_naics5['index'].apply(lambda x: x[1])
nat_naics5['freq'] = nat_naics5['naics5']
nat_naics5['naics5'] = nat_naics5['index'].apply(lambda x: x[4]).astype(str)


nat_naics5 = nat_naics5.reset_index(drop = True)
nat_naics5 = nat_naics5[['year', 'result','naics5', 'freq']]
nat_naics5
```

```python
nat_naics5_join = nat_naics5.groupby(['year', 'result']).apply(pd.DataFrame.sort_values, 'freq', ascending = False).groupby(['year','result']).agg({'naics5': (lambda x: ", ".join(x))})
nat_naics5_join
```

```python
nat_naics5_join['index'] = nat_naics5_join.index
nat_naics5_join['year'] = nat_naics5_join['index'].apply(lambda x: x[0])
nat_naics5_join['result'] = nat_naics5_join['index'].apply(lambda x: x[1])

nat_naics5_join = nat_naics5_join.reset_index(drop = True)
nat_naics5_join = nat_naics5_join[['year', 'result','naics5']]
nat_naics5_join
```

```python
nat_job = dat.groupby(['year', 'case_status'])['job'].value_counts().groupby(['year', 'case_status']).nlargest(5).to_frame()

nat_job['index'] = nat_job.index
nat_job['year'] = nat_job['index'].apply(lambda x: x[0])
nat_job['result'] = nat_job['index'].apply(lambda x: x[1])
nat_job['freq'] = nat_job['job']
nat_job['job'] = nat_job['index'].apply(lambda x: x[4]).astype(str)


nat_job = nat_job.reset_index(drop = True)
nat_job = nat_job[['year', 'result','job', 'freq']]
nat_job
```

```python
nat_job_join = nat_job.groupby(['year', 'result']).apply(pd.DataFrame.sort_values, 'freq', ascending = False).groupby(['year','result']).agg({'job': (lambda x: ", ".join(x))})
nat_job_join
```

```python
nat_job_join['index'] = nat_job_join.index
nat_job_join['year'] = nat_job_join['index'].apply(lambda x: x[0])
nat_job_join['result'] = nat_job_join['index'].apply(lambda x: x[1])

nat_job_join = nat_job_join.reset_index(drop = True)
nat_job_join = nat_job_join[['year', 'result','job']]
nat_job_join
```

#### joining industry and job tables

```python
nat_add1 = pd.merge(nat_naics5_join, nat_job_join, how = 'left', on = ['year', 'result'], validate = "1:1")
nat_add1
```

#### top geographic indicators: employer and worksite zipcodes

```python
dat['emp_citystate'] = dat['emp_city'] + "-" + dat['emp_st']

nat_empcity = dat.groupby(['year', 'case_status'])['emp_citystate'].value_counts().groupby(['year', 'case_status']).nlargest(20).to_frame()

nat_empcity['index'] = nat_empcity.index
nat_empcity['year'] = nat_empcity['index'].apply(lambda x: x[0])
nat_empcity['result'] = nat_empcity['index'].apply(lambda x: x[1])
nat_empcity['freq'] = nat_empcity['emp_citystate']
nat_empcity['empcity'] = nat_empcity['index'].apply(lambda x: x[4]).astype(str)


nat_empcity = nat_empcity.reset_index(drop = True)
nat_empcity = nat_empcity[['year', 'result','empcity', 'freq']]
nat_empcity

nat_empcity_join = nat_empcity.groupby(['year', 'result']).apply(pd.DataFrame.sort_values, 'freq', ascending = False).groupby(['year','result']).agg({'empcity': (lambda x: ", ".join(x))})

nat_empcity_join['index'] = nat_empcity_join.index
nat_empcity_join['year'] = nat_empcity_join['index'].apply(lambda x: x[0])
nat_empcity_join['result'] = nat_empcity_join['index'].apply(lambda x: x[1])

nat_empcity_join = nat_empcity_join.reset_index(drop = True)
nat_empcity_join = nat_empcity_join[['year', 'result','empcity']]

nat_empcity_join
```

```python
dat['work_citystate'] = dat['work_city'] + "-" + dat['work_st']

nat_workcity = dat.groupby(['year', 'case_status'])['work_citystate'].value_counts().groupby(['year', 'case_status']).nlargest(20).to_frame()

nat_workcity['index'] = nat_workcity.index
nat_workcity['year'] = nat_workcity['index'].apply(lambda x: x[0])
nat_workcity['result'] = nat_workcity['index'].apply(lambda x: x[1])
nat_workcity['freq'] = nat_workcity['work_citystate']
nat_workcity['workcity'] = nat_workcity['index'].apply(lambda x: x[4]).astype(str)


nat_workcity = nat_workcity.reset_index(drop = True)
nat_workcity = nat_workcity[['year', 'result','workcity', 'freq']]
nat_workcity

nat_workcity_join = nat_workcity.groupby(['year', 'result']).apply(pd.DataFrame.sort_values, 'freq', ascending = False).groupby(['year','result']).agg({'workcity': (lambda x: ", ".join(x))})

nat_workcity_join['index'] = nat_workcity_join.index
nat_workcity_join['year'] = nat_workcity_join['index'].apply(lambda x: x[0])
nat_workcity_join['result'] = nat_workcity_join['index'].apply(lambda x: x[1])

nat_workcity_join = nat_workcity_join.reset_index(drop = True)
nat_workcity_join = nat_workcity_join[['year', 'result','workcity']]

nat_workcity_join
```

```python
nat_add2 = pd.merge(nat_empcity_join, nat_workcity_join, how = 'left', on = ['year', 'result'], validate = "1:1")
nat_add2
```

```python
nat_samecounty = dat.groupby(['year', 'case_status', 'same_county']).agg({'index': 'count'})
#nat_samest = dat.groupby(['year', 'case_status', 'same_st']).agg({'index': 'count'})

nat_samecounty['freq'] = nat_samecounty['index']
nat_samecounty['index'] = nat_samecounty.index
nat_samecounty['year'] = nat_samecounty['index'].apply(lambda x: x[0])
nat_samecounty['result'] = nat_samecounty['index'].apply(lambda x: x[1])
nat_samecounty['samecounty'] = nat_samecounty['index'].apply(lambda x: x[2])

nat_samecounty = nat_samecounty.reset_index(drop = True)
nat_samecounty = nat_samecounty[['year', 'result','samecounty', 'freq']]
nat_samecounty

nat_samecounty_join = nat_samecounty.groupby(['year', 'result']).apply(pd.DataFrame.sort_values, 'samecounty', ascending = False).groupby(['year','result']).first()

nat_samecounty_join['index'] = nat_samecounty_join.index
nat_samecounty_join['year'] = nat_samecounty_join['index'].apply(lambda x: x[0])
nat_samecounty_join['result'] = nat_samecounty_join['index'].apply(lambda x: x[1])

nat_samecounty_join = nat_samecounty_join.reset_index(drop = True)
nat_samecounty_join = nat_samecounty_join[['year', 'result','freq']]

nat_samecounty_join.columns = ['year', 'result', 'samecounty']
nat_samecounty_join
```

```python
nat_samest = dat.groupby(['year', 'case_status', 'same_st']).agg({'index': 'count'})
#nat_samest = dat.groupby(['year', 'case_status', 'same_st']).agg({'index': 'count'})

nat_samest['freq'] = nat_samest['index']
nat_samest['index'] = nat_samest.index
nat_samest['year'] = nat_samest['index'].apply(lambda x: x[0])
nat_samest['result'] = nat_samest['index'].apply(lambda x: x[1])
nat_samest['samest'] = nat_samest['index'].apply(lambda x: x[2])

nat_samest = nat_samest.reset_index(drop = True)
nat_samest = nat_samest[['year', 'result','samest', 'freq']]
nat_samest

nat_samest_join = nat_samest.groupby(['year', 'result']).apply(pd.DataFrame.sort_values, 'samest', ascending = False).groupby(['year','result']).first()

nat_samest_join['index'] = nat_samest_join.index
nat_samest_join['year'] = nat_samest_join['index'].apply(lambda x: x[0])
nat_samest_join['result'] = nat_samest_join['index'].apply(lambda x: x[1])

nat_samest_join = nat_samest_join.reset_index(drop = True)
nat_samest_join = nat_samest_join[['year', 'result','freq']]

nat_samest_join.columns = ['year', 'result', 'samest']
nat_samest_join
```
