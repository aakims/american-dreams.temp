```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
data = pd.read_csv("Austin_Animal_Center_Intakes.csv", dtype = {'Name': str, 'Found Location': str, 'Age upon Intake': str, 'Breed': str, 'Color': str})
```

```python
df = data.rename(columns = {'Animal ID': 'id', 'Name': 'name', 'DateTime': 'found_time', 'Found Location': 'found_location', 'Intake Type': 'intake_type', 'Intake Condition': 'intake_cond', 'Animal Type': 'animal_type', 'Sex upon Intake': 'sex_intake', 'Age upon Intake': 'age_intake', 'Breed': 'breed', 'Color': 'color'})
```

```python
df['intake_type'] = df.intake_type.astype('category')
df['intake_cond'] = df.intake_cond.astype('category')
df['animal_type'] = df.animal_type.astype('category')
df['sex_intake'] = df.sex_intake.astype('category')
```

```python
df.name = df.name.fillna('NoName')
```

```python
from dateutil.parser import parse
df['found_time'] = df.found_time.astype(str)
```

```python
df['found_time'] = pd.to_datetime(df['found_time'])
```

```python
df2014= df[(df['found_time'] > '2013-12-31') & (df['found_time'] < '2015-01-01')]
```

```python
len(df2014)
```

```python
df2017 = df[(df['found_time']>'2016-12-31') & (df['found_time'] < '2018-01-01')]
```

```python
len(df2017)
df2017
```

```python
df2017 = df2017.drop(['MonthYear'], axis = 1)
```

```python
df2014 = df2014.drop(['MonthYear'], axis = 1)
```

```python
df17c = df2017
```

```python
df17c['addressbits'] = df17c['found_location'].str.split(' ')
```

```python
df17c.commas = df17c['addressbits'].apply(lambda x: len(x))
```

```python
df17Geo = df17c[df17c.commas != 2]
```

```python
df17Geo['state'] = df17Geo['addressbits'].apply(lambda x: x[len(x)-1])
```

```python
df17Geo['state'] = df17Geo['state'].str.extract(r"\((.*?)\)", expand=False)
```

```python
df17Geo['city'] = df17Geo['addressbits'].apply(lambda x: x[len(x) - 2])
```

```python
df17Geo['address'] = df17Geo['addressbits'].apply(lambda x: x[0:(len(x)-3)])
```

```python
df17Geo['address'] = df17Geo['address'].apply(' '.join)
```

```python
df17GeoExport = df17Geo.drop(['found_location', 'commas', 'addressbits'], axis = 1)
```

```python
len(df17GeoExport)
```

```python
df17GeoExport.to_csv("2017Intakes.csv")
```
