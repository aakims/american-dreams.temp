### Separate agg tables then merging approach

It seems hard to create multiple nlargest aggregation columns with everything else coming before it. I'm going to create multiple groupby (grouping using the same columns) tables then merge their results later

### I. Set up the index and basic numerical stats

```python
## basic setup

nat_group = ['year', 'quarter','months','case_status']
nat_agg = {
    'index': 'count', 
    'tot_workers': np.sum, 
    'yr_sal': [np.min, np.mean, np.max],
    'pw_sal': [np.min, np.mean, np.max],
    'dec_dur': (lambda x: x.mean())
}
```

#### groupby function

```python
nat_byYear = dat.groupby(['year', 'case_status']).agg(nat_agg)

nat_byYear
```

#### clean up the tabular format

```python
## Renaming columns 
nat_byYear.columns = ["_".join(x) for x in nat_byYear.columns.ravel()]
nat_byYear.columns = ['app_count', 'worker_count', 
                      'min_yr_sal', 'avg_yr_sal', 'max_yr_sal', 
                      'min_pw', 'avg_pw', 'max_pw', 
                      'avg_decdur']
```

```python
nat_byYear['index'] = nat_byYear.index
nat_byYear['year'] = nat_byYear['index'].apply(lambda x: x[0])
nat_byYear['result'] = nat_byYear['index'].apply(lambda x: x[1])
##nat_byYear['peiord'] = nat_byYear['index'].apply(lambda x: x[2])
```

```python
nat_byYear.head()
```

```python
# float values to int 
nat_byYear['min_yr_sal'] = nat_byYear['min_yr_sal'].apply(np.floor).astype(int)
nat_byYear['avg_yr_sal'] = nat_byYear['avg_yr_sal'].apply(np.floor).astype(int)
nat_byYear['max_yr_sal'] = nat_byYear['max_yr_sal'].apply(np.floor).astype(int)
nat_byYear['min_pw'] = nat_byYear['min_pw'].apply(np.floor).astype(int)
nat_byYear['avg_pw'] = nat_byYear['avg_pw'].apply(np.floor).astype(int)
nat_byYear['max_pw'] = nat_byYear['max_pw'].apply(np.floor).astype(int)

# round to 'days'
nat_byYear['avg_decdur'] = nat_byYear['avg_decdur'].dt.days
```
