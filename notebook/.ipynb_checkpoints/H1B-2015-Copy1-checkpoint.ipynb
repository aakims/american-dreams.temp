```python
#2015 ONLY
vdata.loc[(vdata['WAGESPLIT'].apply(lambda x: len(x) == 3)),'WAGERANGE'] = 'YES'
vdata.loc[(vdata['WAGESPLIT'].apply(lambda x: len(x) == 2)),'WAGERANGE'] = 'NO'
vdata['WAGE_FROM'] = vdata['WAGESPLIT'].apply(lambda x: x[0])
vdata.loc[(vdata['WAGERANGE'] == 'YES'), 'WAGE_TO'] = vdata['WAGESPLIT'].apply(lambda x: x[len(x)-1])
```

```python
vdata['WAGE_FROM'] = vdata['WAGE_FROM'].astype(float)
```

```python
### IF above throws an error, use below 

## because there were values found with '-' which cannot be converted into float values
# i'm looking for cells that has '-' 
vdata[vdata['WAGE_FROM'].apply(lambda x: any(pd.Series(x).str.contains('-')))]
```

```python
## selective data cleaning 2015
# 22.69-61.55, and rest just '-'
vdata.loc[(vdata['WAGE_FROM'] == '22.69-61.55'),'WAGE_TO'] = '61.55' #[22.69-61.55, -, 46.02]
vdata.loc[(vdata['WAGE_FROM'] == '22.69-61.55'),'WAGE_FROM'] = '22.69'
vdata.loc[(vdata['WAGE_FROM'] == '-'),'WAGE_FROM'] = 'NaN'
```

```python
## more cleaning 
vdata.loc[(vdata['WAGE_TO'] == 'N/A'), 'WAGE_TO'] = 'NaN'
vdata.loc[(vdata['WAGE_TO'] == '+'), 'WAGE_TO'] = 'NaN'
vdata.loc[(vdata['WAGE_TO'] == '-'), 'WAGE_TO'] = 'NaN'
vdata.loc[(vdata['WAGE_TO'].notnull()), 'WAGE_MEMO'] = vdata['WAGE_TO'].astype(float)

#########
#### now re-run: vdatafull['WAGE_FROM'] = vdatafull['WAGE_FROM'].astype(float)
```

```python
vdata['WAGE_FROM'] = vdata['WAGE_FROM'].astype(float)
vdata['WAGE_TO'] = vdata['WAGE_TO'].astype(float)
vdata['WAGE'] = vdata['WAGE_FROM']
vdata.loc[(np.isfinite(vdata['WAGE_TO'])),'WAGE'] = 0.5* (vdata['WAGE_FROM'] + vdata['WAGE_TO'])
```

```python
### WHEN EMEPLOYER ZIPCODE AVAILABLE 
#join/merging to get clean state fips of employer for each records
vdataem = pd.merge(vdata, georefz, how = 'left', left_on='EMPLOYER_POSTAL_CODE', right_on='zip', validate="m:m")
vdataem = vdataem.rename(columns = {'zip': 'EMP_ZIP', 'county': 'EMP_CNTY'})
```

```python
vdataem = vdataem.drop_duplicates(subset=['ROW_ID'], keep = 'first') 
len(vdataem)
```

```python
## WHEN WORKSITE ZIPCODE AVAILABLE 
vdataewm = pd.merge(vdataem, georefz, how = 'left', left_on=['WORKSITE_POSTAL_CODE'] , right_on='zip', validate="m:m")
vdatafull = vdataewm.rename(columns = {'zip': 'WORK_ZIP', 'county': 'WORK_CNTY'})
```

```python
vdatafull.loc[(vdatafull['CASE_STATUS'] =='DENIED'), 'RESULT'] = 'DENIED'
vdatafull.loc[(vdatafull['CASE_STATUS'] =='CERTIFIED'), 'RESULT'] = 'CERTIFIED'
vdatafull.loc[(vdatafull['CASE_STATUS'] =='CERTIFIED-WITHDRAWN'), 'RESULT'] = 'CERT-WITHD'
vdatafull.loc[(vdatafull['CASE_STATUS'] =='WITHDRAWN'), 'RESULT'] = 'WITHDRAWN'
```

```python
#normaling wages 
#visa17all['PW_UNIT_OF_PAY'].unique() 
# [Year, Hour, Month, NaN, Week, Bi-Weekly]
# hourly: x40 x52
# Month: x12
# Week: x52
# Bi-Weekly: x26
vdatafull.loc[(vdatafull['PAYUNIT'] =='Year'),'YrSalary'] = vdatafull['WAGE']
```

```python
vdatafull.loc[(vdatafull['PAYUNIT'] =='Month'),'YrSalary'] = vdatafull['WAGE']*12
```

```python
vdatafull.loc[(vdatafull['PAYUNIT'] =='Hour'),'YrSalary'] = vdatafull['WAGE']*40*52
```

```python
vdatafull.loc[(vdatafull['PAYUNIT'] =='Week'),'YrSalary'] = vdatafull['WAGE']*52
```

```python
vdatafull.loc[(vdatafull['PAYUNIT'] =='Bi-Weekly'),'YrSalary'] = vdatafull['WAGE']*26
```

```python
vdatafull['YrSalary'] = vdatafull['YrSalary'].astype(float)
```

```python
employerLoc = ts.groupby(['EMP_CNTY','RESULT','PERIOD']).agg({'TOTAL_WORKERS': np.sum, 'YrSalary': [np.sum, np.mean], 'DEC_DUR': (lambda x: x.mean()), 'NAICS4': (lambda x: x.value_counts().index[0])})
```

```python
worksiteLoc = ts.groupby(['WORK_CNTY','RESULT','PERIOD']).agg({'TOTAL_WORKERS': np.sum, 'YrSalary': [np.sum, np.mean], 'DEC_DUR': (lambda x: x.mean()), 'NAICS4': (lambda x: x.value_counts().index[0])})
```

```python
employerLoc.columns = ["_".join(x) for x in employerLoc.columns.ravel()]
worksiteLoc.columns = ["_".join(x) for x in worksiteLoc.columns.ravel()]
```

```python
worksiteLoc.head()
```

```python
employerLoc = employerLoc.rename(columns = {'TOTAL_WORKERS_sum' : 'tot_workers', 'YrSalary_sum': 'all_salary', 'YrSalary_mean': 'avg_salary', 'DEC_DUR_<lambda>': 'decision_dur', 'NAICS4_<lambda>': 'naics4'})
worksiteLoc = worksiteLoc.rename(columns = {'TOTAL_WORKERS_sum' : 'tot_workers', 'YrSalary_sum': 'all_salary', 'YrSalary_mean': 'avg_salary', 'DEC_DUR_<lambda>': 'decision_dur','NAICS4_<lambda>': 'naics4'})
```

```python
employerLoc['index'] = employerLoc.index
worksiteLoc['index'] = worksiteLoc.index
```

```python
worksiteLoc['cfips'] = worksiteLoc['index'].apply(lambda x: x[0])
worksiteLoc['result'] = worksiteLoc['index'].apply(lambda x: x[1])
worksiteLoc['peiord'] = worksiteLoc['index'].apply(lambda x: x[2])
```

```python
employerLoc['cfips'] = employerLoc['index'].apply(lambda x: x[0])
employerLoc['result'] = employerLoc['index'].apply(lambda x: x[1])
employerLoc['peiord'] = employerLoc['index'].apply(lambda x: x[2])
```

```python
#list(h1bbyWorkCounty) 
#h1bbyWorkCounty['agg_salary'].apply(np.floor)
```

```python
worksiteLoc['avg_salary'] = worksiteLoc['avg_salary'].apply(np.floor)
employerLoc['avg_salary'] = employerLoc['avg_salary'].apply(np.floor)
```

```python
worksiteLoc['all_salary'] = worksiteLoc['all_salary'].apply(np.floor)
employerLoc['all_salary'] = employerLoc['all_salary'].apply(np.floor)
```

```python
worksiteLoc = worksiteLoc[['cfips','result','peiord', 'tot_workers',
 'all_salary','avg_salary', 'decision_dur','naics4']]
employerLoc = employerLoc[['cfips','result','peiord', 'tot_workers',
 'all_salary','avg_salary', 'decision_dur', 'naics4']]
```

```python
worksiteLoc = worksiteLoc.rename(columns = {'peiord': 'period'})
employerLoc = employerLoc.rename(columns = {'peiord': 'period'})
```

```python
worksiteLoc['decision_dur'] = (worksiteLoc['decision_dur']).dt.days
employerLoc['decision_dur'] = (employerLoc['decision_dur']).dt.days
```
